```python
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
limited_wr = pd.read_csv('../Data/limited_wr_card_names.csv')
limited_wr
```

```python
standard_wr = pd.read_csv('../Data/standard_wr_card_names.csv')
standard_wr
```

```python
std_lim_wr = pd.merge(limited_wr, standard_wr, on='card', how='inner')
std_lim_wr.sort_values(by = 'std_win_rate', ascending = False)
```

```python
ax = sns.regplot(data=std_lim_wr, x="std_win_rate", y="lim_win_rate")
ax.annotate('Terror Tide', xy=(0.468, 0.556482), xytext=(.47, .61), fontsize = 8 ,
            arrowprops=dict(facecolor='black', shrink=0.05, width = 2, headwidth = 6))
ax.annotate('Fabrication Foundry', xy=(0.493, 0.475342), xytext=(.51, .48), fontsize = 8 ,
            arrowprops=dict(facecolor='black', shrink=0.05, width = 2, headwidth = 6))
ax.annotate('Bonehoard Dracosaur', xy=(0.518, 0.705458), xytext=(.48, .69), fontsize = 8 ,
            arrowprops=dict(facecolor='black', shrink=0.05, width = 2, headwidth = 6))
ax.annotate('Sanguine Evangelist', xy=(0.563, 0.653733), xytext=(.535, .67), fontsize = 8 ,
            arrowprops=dict(facecolor='black', shrink=0.05, width = 2, headwidth = 6))
ax.set(ylabel='Limited Win Rate', xlabel='Standard Win Rate')
plt.title("Successful Standard Cards' Performance in Limited.")
plt.savefig('../Data/std_lim_wr.png')
```
